```python
import collections  #This imports the collections module, which provides
#specialized container datatypes beyond the built-in containers like list, dict, and tuple.

import pandas as pd

import networkx as nx
#This imports the NetworkX library and assigns it the alias nx. NetworkX is a Python package for the creation,
#manipulation, and study of the structure, dynamics, and functions of complex networks.

import matplotlib.pyplot as plt
```

```python
# code for plotting hypothetical LinkedIn network
df = pd.DataFrame([("Dave", "Jenny"), ("Peter", "Jenny"), ("John", "Jenny"),
                   ("Dave", "Peter"), ("Dave", "John"), ("Peter", "Sam"),
                   ("Sam", "Albert"), ("Peter", "John")], columns=['from', 'to'])
G = nx.from_pandas_edgelist(df, 'from', 'to')
nx.draw(G, with_labels=True, node_color='skyblue')
plt.show()

#the DataFrame contains a list of tuples representing edges in a graph. Each tuple consists of two nodes 
#connected by an edge. The columns of the  DataFrame are named 'from' and 'to', representing 
#the source and target nodes of each edge, respectively.
#This line creates a graph G using NetworkX's from_pandas_edgelist function. It takes the DataFrame df as input 
#and specifies which columns represent the source and target nodes ('from' and 'to' columns, respectively).
#This line visualizes the graph G using NetworkX's draw function. It draws the nodes and edges of the graph. 
#The with_labels=True argument indicates that node labels should be displayed, and node_color='skyblue' sets 
#the color of the nodes to sky blue
```

# Directed vs Undirected Network

```python
# use nx.DiGraph to create a directed network
#This modified code snippet creates a directed graph (also known as a directed network)
#instead of an undirected graph.
G = nx.from_pandas_edgelist(df, 'from', 'to', create_using=nx.DiGraph()) 
nx.draw(G, with_labels=True, node_color='skyblue')
plt.show()

#By specifying create_using=nx.DiGraph(), it indicates that the graph should be a directed graph (DiGraph). 
#This means that edges have a direction, going from one node (the source) to another node (the target).
```

# Visualizing and Analyzing Networks- plot layout

```python
# code for plotting different layouts
G = nx.from_pandas_edgelist(df, 'from', 'to')
plt.subplot(121) 
nx.draw_circular(G, with_labels=True, node_color='skyblue', node_size=1000)
plt.subplot(122)
nx.draw_kamada_kawai(G, with_labels=True, node_color='skyblue', node_size=1000)
plt.show()

#plt.subplot(121): This line sets up a subplot grid with 1 row and 2 columns 
#and activates the first subplot (index 1). The 121 argument specifies that there should be 1 row, 2 columns, 
#and the current subplot being activated is the first one.
#plt.subplot(122): This line activates the second subplot (index 2) in the same subplot grid. 
#The 122 argument specifies that the current subplot being activated is the second one.
#nx.draw_circular(G, with_labels=True, node_color='skyblue', node_size=1000): This line draws 
#the graph G using NetworkX's draw_circular function in the first subplot. It arranges the nodes 
#in a circular layout, where nodes are positioned evenly around a circle. The with_labels=True argument 
#indicates that node labels should be displayed, and node_color='skyblue' sets the color of the nodes to sky blue.
#node_size=1000 sets the size of the nodes.
#nx.draw_kamada_kawai(G, with_labels=True, node_color='skyblue', node_size=1000): This line draws the graph G 
#using NetworkX's draw_kamada_kawai function in the second subplot. It arranges the nodes using the Kamada-Kawai 
#force-directed layout algorithm, which optimizes the positions of nodes to minimize the stress of the layout. 
```

# EgoCentric Network

```python
# nx.ego_graph get Peter’s 1-level ego network
# for a 2-level ego network set argument order = 2 in make_ego_graph().
plt.subplot(121)
# get 1-level ego network for ’Peter’
G_ego = nx.ego_graph(G, 'Peter')
nx.draw(G_ego, with_labels=True, node_color='skyblue')
plt.subplot(122)
# set radius=2 to get 2-level ego network
G_ego = nx.ego_graph(G, 'Peter', radius=2)
nx.draw(G_ego, with_labels=True, node_color='skyblue')

#G_ego = nx.ego_graph(G, 'Peter'): This line creates a 1-level ego network for Peter in the graph G. It includes 
#Peter and all nodes connected to Peter within one hop (directly connected nodes).
#G_ego = nx.ego_graph(G, 'Peter', radius=2): This line creates a 2-level ego network for Peter in the graph G. 
#It includes Peter, all nodes connected to Peter within one hop, and all nodes connected to those nodes within 
#two hops.
```
